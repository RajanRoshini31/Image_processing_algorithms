```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
noisy_image=cv2.imread('/content/drive/MyDrive/IP/lena.jfif',cv2.IMREAD_GRAYSCALE)
restored_image=cv2.fastNlMeansDenoising(noisy_image,None,h=10,templateWindowSize=7,searchWindowSize=21)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Noisy Image')
plt.imshow(noisy_image,cmap='gray')
plt.subplot(1,2,2)
plt.title('Restored Image(Denoised)')
plt.imshow(restored_image,cmap='gray')
plt.show()
```

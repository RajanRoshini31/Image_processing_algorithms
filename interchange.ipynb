```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img =cv.imread('/content/download.jfif')
```

```python
b,g,r=cv.split(img)
```

```python
img_interchanged =cv.merge((r,g,b))
```

```python
plt.figure(figsize=(5,5))
plt.subplot(121)
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.title("Original Image")
plt.subplot(122)
plt.imshow(cv.cvtColor(img_interchanged,cv.COLOR_BGR2RGB))
plt.title("Red and Blue Planes InterChanged")
plt.show()
```

```python
img =cv.imread('/content/download.jfif')
b,g,r=cv.split(img)
img_interchanged =cv.merge((r,g,b))
plt.figure(figsize=(5,5))
plt.subplot(212)
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.title("Original Image")
plt.subplot(211)
plt.imshow(cv.cvtColor(img_interchanged,cv.COLOR_BGR2RGB))
plt.title("Red and Blue Planes InterChanged")
plt.show()
```

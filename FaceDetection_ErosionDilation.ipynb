```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
reference_image =cv2.imread('/content/images12.jpg',0);
target_image =cv2.imread('/content/images123.jpg',0);
orb=cv2.ORB_create();
keypoints_ref,descriptions_ref=orb.detectAndCompute(reference_image,None);
keypoints_target,descriptions_target=orb.detectAndCompute(target_image,None);
bf=cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True);
matches=bf.match(descriptions_ref,descriptions_target);
matches=sorted(matches,key=lambda x:x.distance);
f=cv2.DrawMatchesFlags_NOT_DRAW_SINGLE_POINTS
result_image=cv2.drawMatches(reference_image,keypoints_ref,target_image,keypoints_target,matches[:50],None,flags=f);
plt.figure(figsize=(10,6))
plt.imshow(result_image);
plt.title("Face Detetction using Feature Matching")
plt.show();
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image =cv2.imread('/content/drive/MyDrive/IP/lena.jfif',0)
_,binary_image=cv2.threshold(image,127,255,cv2.THRESH_BINARY)
kernel =np.ones((3,3),np.uint8)
erosion =cv2.erode(binary_image,kernel,iterations=1)
dilation =cv2.dilate(binary_image,kernel,iterations=1)
titles=['Orginal Image','Erosion','Dilation']
images=[image,erosion,dilation]
for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

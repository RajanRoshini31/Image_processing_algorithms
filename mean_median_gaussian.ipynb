```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def display_image_and_histogram(image, title):
    plt.figure(figsize=(12, 6))

    # Display image
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f'{title} Image')
    plt.axis('off')

    # Display histogram
    plt.subplot(1, 2, 2)
    plt.hist(image.flatten(),256,[0,256],color='r')
    plt.title(f'{title} Histogram')
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()
image = cv2.imread('/content/drive/MyDrive/IP/lena.jfif',cv2.IMREAD_GRAYSCALE)

# Apply Mean Filter
mean_filtered = cv2.blur(image, (5, 5))

# Apply Median Filter
median_filtered = cv2.medianBlur(image, 5)

# Apply Gaussian Filter
gaussian_filtered = cv2.GaussianBlur(image, (5, 5), 0)

# Display results
display_image_and_histogram(image, 'Original')
display_image_and_histogram(mean_filtered, 'Mean Filtered')
display_image_and_histogram(median_filtered, 'Median Filtered')
display_image_and_histogram(gaussian_filtered, 'Gaussian Filtered')
```

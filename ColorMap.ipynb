```python
import matplotlib.pyplot as plt
import numpy as np
print(plt.colormaps())
x=plt.colormaps()
for i in x:
  data = np.random.rand(10, 10)
  plt.imshow(data, cmap=i, interpolation='nearest')
  plt.colorbar()
  plt.title('Heatmap with '+i)
  plt.show()
```

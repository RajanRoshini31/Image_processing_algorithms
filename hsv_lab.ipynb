```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_image(image_path):
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image_rgb
def convert_to_hsv(image_rgb):
    image_hsv = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)
    return image_hsv
def convert_to_lab(image_rgb):
    image_lab= cv2.cvtColor(image_rgb, cv2.COLOR_RGB2LAB)
    return image_lab
def plot_image(image,title,color_map=None):
  plt.imshow(image,cmap=color_map)
  plt.title(title)
  plt.axis('off')
def analyse_image(image_path):
  image_rgb =load_image(image_path)
  image_hsv = convert_to_hsv(image_rgb)
  image_lab = convert_to_lab(image_rgb)
  plt.figure(figsize=(18,6))
  plt.subplot(1,3,1)
  plot_image(image_rgb,'RGB Color Model')

  plt.subplot(1,3,2)
  plot_image(image_hsv,'HSV Color Model')

  plt.subplot(1,3,3)
  plot_image(image_lab,'LAB Color Model')
  plt.show()

image_path ='/content/lena.jfif'
analyse_image(image_path)
image_path ='/content/flower.jpg'
analyse_image(image_path)
```
